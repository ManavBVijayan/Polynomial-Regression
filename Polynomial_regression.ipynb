```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class PolynomialRegression:
    def __init__(self, degree):
        self.degree = degree
        self.coefficients = None

    def fit(self, X, Y):
        # Create the feature matrix with polynomial terms
        X_poly = np.column_stack([X**i for i in range(1, self.degree + 1)])

        # Solve for coefficients using the normal equation
        self.coefficients = np.linalg.inv(X_poly.T @ X_poly) @ X_poly.T @ Y

    def predict(self, X):
        # Create the feature matrix for prediction
        X_poly = np.column_stack([X**i for i in range(1, self.degree + 1)])

        # Make predictions
        Y_pred = X_poly @ self.coefficients

        return Y_pred
```

```python
# Generate sample data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 8, 12, 20, 30])
```

```python
# Choose the degree of the polynomial
degree = 2
```

```python
# Create and train the polynomial regression model
poly_regression = PolynomialRegression(degree)
poly_regression.fit(X, Y)
```

```python
# Make predictions for new data points
X_pred = np.linspace(min(X), max(X), 100)
Y_pred = poly_regression.predict(X_pred)
```

```python
# Plot the original data points and the polynomial regression curve
plt.scatter(X, Y, label='Actual Data')
plt.plot(X_pred, Y_pred, color='red', label='Polynomial Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
